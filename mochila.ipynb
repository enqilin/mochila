```python
import pandas as pd
```

```python
df = pd.read_csv('../mochila/pokemon.csv')
df.info()
```

```python
df['total_stadistics'] = df['attack']+df['hp']+df['sp_attack']+df['sp_defense']+df['defense']+df['speed']
df['capture_rate'] = df['capture_rate'].drop(df[df['capture_rate'] == '30 (Meteorite)255 (Core)'].index)
df['capture_rate'].fillna(0, inplace=True)
df['capture_rate'] = df['capture_rate'].astype(int)
df['capture_rate'].info()
```

```python
df['peso'] =1 - df['capture_rate'] / df['capture_rate'].max()
df['stadistics'] = 1-(df['total_stadistics'] / df['total_stadistics'].max())

df['media'] = df['stadistics'] / df['peso']
df
```

```python
columns= ['name','type1','type2','generation','attack','hp','sp_attack','sp_defense','defense','speed','total_stadistics','capture_rate']
df_2=df[columns]
df_2
```

```python
columns_1 = ['name','peso','stadistics','media']
df_3=df[columns_1]
df_3.fillna(0, inplace=True)
```

```python
df_2.to_csv('../mochila/pokemon_2.csv', index=False)
df_3.to_csv('../mochila/pokemon_no_normalizar.csv', index=False)
```

```python
def es_solucion(solucion, capacidad):
    """
    Determina si el conjunto solucion es una solucion valida al problema
    """
    total = 0 
    for elemento in solucion:
        total = round(total + (elemento[1], 2))
        return total <= capacidad
    if total == capacidad:
        return True
    else:
        return False
    
def mochila(conjunto_candidatos, capacidad):
    solucion = []
    restante = capacidad
    while conjunto_candidatos and not es_solucion(solucion, capacidad):
        dato = conjunto_candidatos.pop(0)
        if dato[1] <= restante:
            solucion.append(dato)
            restante  = round(restante - dato[1], 2)
    return solucion
```
